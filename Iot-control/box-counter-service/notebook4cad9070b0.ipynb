```python
from ultralytics import YOLO
import os
```

```python
dataset_path = "/kaggle/input/datasets/tuyennguyen21pt/dataset"

for root, dirs, files in os.walk(dataset_path):
    print(root, len(files))
```

```python
model = YOLO("yolo11n.pt")
```

```python
model.train(
    data="/kaggle/input/datasets/tuyennguyen21pt/dataset/dataset/data.yaml",          # hoặc data_yaml
    epochs=100,
    imgsz=640,
    batch=16,
    workers=2,
    device=0,
    project="box_counter",
    name="yolo11"
)
```

```python
metrics = model.val()
print(metrics)
```

```python
results = model.predict(
    source="/kaggle/input/datasets/tuyennguyen21pt/tessttst",
    conf=0.25,
    save=True
)
```

```python
results = model.predict(
    "/kaggle/input/datasets/tuyennguyen21pt/tessttst"
)

count = len(results[0].boxes)

print("Number of boxes:", count)
```
